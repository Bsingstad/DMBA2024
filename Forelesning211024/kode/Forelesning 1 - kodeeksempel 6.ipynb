```python
import numpy as np
import os
import math
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
%matplotlib inline
import random

from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
```

```python
def f(x, y):
    return x*x + 2*y*y

# Make data set
x = np.linspace(-10, 10, 30)
y = np.linspace(-10, 10, 30)

x_mesh, y_mesh = np.meshgrid(x, y)
z = f(x_mesh, y_mesh)
```

```python
x_0 = random.randint(-10,10)
y_0 = random.randint(-10,10)
z_0 = f(x_0, y_0) + 25
```

```python
def delta_f(x,y):
    return 2*x, 4 * y
```

```python
#x_0 = random.randint(-10,10)
x_0 = -10
#y_0 = random.randint(-10,10)
y_0 = -10
z_0 = f(x_0, y_0)

lr = 0.1
iterations = 10
z_log = []
x_log = []
y_log = []
z_log.append(z_0)
y_log.append(y_0)
x_log.append(x_0)

for i in range(iterations):
    x_temp, y_temp = delta_f(x_0,y_0)
    x_updated = x_0 - x_temp * lr
    y_updated = y_0 - y_temp * lr
    z_updated = f(x_updated, y_updated)
    z_log.append(z_updated)
    y_log.append(y_updated)
    x_log.append(x_updated)
    x_0 = x_updated
    y_0 = y_updated

ax = plt.axes(projection='3d')
ax.plot(x_log, y_log, z_log, "-r", lw=3, zorder=3, linestyle='--', marker='o')
ax.plot_surface(x_mesh, y_mesh, z, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')
ax.view_init(45, 35)

ax.set_title('surface');
```
