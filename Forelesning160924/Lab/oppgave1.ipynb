```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_tot = pd.read_csv("../data/Titanic/tot_pop.csv")
```

```python
df_sub = pd.read_csv("../data/Titanic/sub_pop.csv")
```

```python
df_tot = df_tot.drop(columns="Survived")
```

#### 1.Visualiser kjønnsfordelingen i totalpopulasjon og utvalget i en og samme figur

```python
fig, ax = plt.subplots(1, 2)

# Teller antall male og female i Sex kolonnen i et utvalg av poulasjonen
male_count_sub = df_sub["Sex"].value_counts()["male"]
female_count_sub = df_sub["Sex"].value_counts()["female"]

# Teller antall male og female i Sex kolonnen i hele populasjonen 
male_count_tot = df_tot["Sex"].value_counts()["male"]
female_count_tot = df_tot["Sex"].value_counts()["female"]

sex_counts_sub = [male_count_sub, female_count_sub]
sex_counts_tot = [male_count_tot, female_count_tot]

# Lager to bar charts 
ax[0].bar(["Mann", "Kvinne"], sex_counts_sub, color=["lightblue", "pink"])
ax[1].bar(["Mann", "Kvinne"], sex_counts_tot, color=["lightblue", "pink"])

# ax[0].hist(df_sub["Sex"], bins=2, rwidth=0.7, )
ax[0].set_title("Utvalg kjønnsfordeling")
ax[0].set_ylabel("Antall")
ax[0].set_xlabel("Kjønn")

# ax[1].hist(df_tot["Sex"], bins=2, rwidth=0.7)
ax[1].set_title("Total kjønnsfordeling")
# ax[1].set_ylabel("Antall")
ax[1].set_xlabel("Kjønn") 

# Alternativ plott
fig2, ax2 = plt.subplots()
ax2.bar(["Mann", "Kvinne"], [male_count_tot, female_count_tot], bottom=[male_count_sub, female_count_sub], 
        color=["lightblue", "pink", "purple", "yellow"])
ax2.set_title("Fordeling av kjønn i et utvalg og hele poulasjonen")
```
