```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
my = 0 # gjennomsnitt
sigma = 0.5 # standardavvik
for i in np.linspace(0,2,11):
    norm_dist_a = np.random.normal(my, sigma, 100)
    norm_dist_b = np.random.normal(my + i, sigma, 100)

    sns.histplot(data=norm_dist_a, kde=True, bins=20, color="C0", label="dist a")
    sns.histplot(data=norm_dist_b, kde=True, bins=20, color="C1", label="dist b")
    _,pval = scipy.stats.ttest_ind(norm_dist_a,norm_dist_b, equal_var=True, axis=0, alternative="two-sided")
    #_,pval = scipy.stats.ttest_ind(norm_dist_a,norm_dist_b, equal_var=True, axis=0, alternative="less")
    plt.title(f"Er distribusjon A signifikant forskjellig fra distribusjon B? P-value={round(pval,10)}")
    plt.legend()
    plt.show()
```
