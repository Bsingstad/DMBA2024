# Obligatorisk innlevering 2 - oppgave 1

Denne oppgaven tar for seg et datasett fra Food Mart, som er ennærbutikkjede i USA. Food Mart har i dette datasettet samlet endel data i forbindelse med mediekampanjer. Din oppgave er å forutsi kostnadene for mediekampanjer i matvaremarkedet på grunnlag av funksjonene som tilbys.

```python
# Importer resten av pakkene du trenger her:
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
import seaborn as sns
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import cross_validate, KFold
```

```python
# I denne oppgaven skal dere bruke følgende datasett:
df = pd.read_csv("./datasett1.csv")
# Her deler jeg datasettet inn i egenskaper (X) og target (y)
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
# Tilslutt deler jeg inn i et trening- og et testsett (75%/25%)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=0)
# PS: Ikke endre denne kodecellen!
```

#### Oppgave 1.2) 

fullfør funksjonen under slik at den tar inn både egenskapene og target, legger disse sammen i samme matrise igjen og visualiser dataene som en korrelasjonsmatrise (poenget er at du skal kunne se hvordan egenskapene korrelerer med target).

* Beskriv hvilke egenskaper du ser korrelerer mest med target

```python
def korrelasjonsmatrise(X,y):
    X_temp = X.copy()
    y_temp = y.copy()
    X_temp["Target"] = y_temp
    #TODO: Fullfør koden
    correlation = X_temp.corr()
    sns.heatmap(correlation)
    plt.title('Korrelasjonsmatrise av dataen')
```

```python
korrelasjonsmatrise(X_train,y_train)
```

**Svar:**

Det er egenskapene video_store, salad_bar, prepared_food og florist som korrelerer mest med target.

#### Oppgave 1.3) 
Velg en modellarkitektur (gjerne fra Scikit-Learn) og gjør et strukturert hyperparametersøk. 
* Gjør et søk på minst 3 paramtere og minst 2 verdier pr parameter
* Bruk 3-folds kryssvalidering for hver parameterkombinasjon
* Velg en metrikk som passer for målet (tips: "neg_mean_squared_error" i sklearn) 
* Skriv ut den parameterkobinasjonen som ga den beste skåren, samt skåren. Kommenter resultatet

```python
#TODO: Utfør hyperparametersøk for en modell
params = {
    'max_depth': [None, 10, 20, 30, 40, 50],
    'min_samples_split': [2, 5, 10, 20],
    'min_samples_leaf': [1, 2, 5, 10]
}

# Bruker DecisionTreeRegressor og Grid Search
model = DecisionTreeRegressor(random_state = 42)
grid = GridSearchCV(estimator = model, 
                    param_grid = params,
                    scoring = 'neg_mean_squared_error',
                    cv = 3)

grid.fit(X_train, y_train)
best_param = grid.best_params_
best_score = grid.best_score_

print(f'Beste parametere: {best_param}')
print(f'Beste score: {best_score:.4f}') 
```

**Svar:**

Ved kjøring er de beste parameterene max_depth = 10, min_samples_leaf = 10 og min_samples_split = 2.
Scoren blir -904.5, som ikke er en så bra score siden man ønsker at scoren skal være så nærme 0 så mulig. 

#### Oppgave 1.4) 
Bruk hyperparameterkombinasjonen fra oppgave 1.3 og kryssvalider den modellen på nytt med K >= 5 
* Logg "mean absolute error" og "mean squared error" for hver kryssvalideringsrunde
* Oppgi gjennomsnitt og standardavvik for "mean absolute error" og "mean squared error" etter kryssvalideringen og kommenter resultatet
* Hvor stor er den gjennomsnittlige feilen på prediksjonen sammenlignet med gjennomsnittlig faktisk pris i treningsdataene? 

```python
#TODO: Utfør kryssvalidering med de optimale parameterene fra forrige oppgave og logg mean_absolute_error og mean_squared_error
# Oppretter modell med de beste parameterene fra tidligere oppgave
opt_model = DecisionTreeRegressor(
    random_state = 42,
    max_depth = 10,
    min_samples_split = 2,
    min_samples_leaf = 10)

# Utfører kryssvalidering med 5 runder
cv = KFold(n_splits = 5, shuffle = True, random_state = 42)
scoring = ['neg_mean_absolute_error', 'neg_mean_squared_error']
cv_results = cross_validate(opt_model, X_train, y_train, cv = cv, scoring = scoring)

cv_mean_absolute_error = cv_results['test_neg_mean_absolute_error']
cv_mean_square_error = cv_results['test_neg_mean_squared_error']

# Logg mean absolute error og mean squared error for hver runde
print('Mean absolute error:')
for i, score in enumerate(cv_mean_absolute_error):
    print(f'Runde {i}: {score:.2f}')
    
print('\nMean square error:')
for i, score in enumerate(cv_mean_square_error):
    print(f'Runde {i}: {score:.2f}')
```

```python
#TODO: Beregn gjennomsnitt og standard avvik for mean_absolute_error og mean_squared_error og beregn hvor stor den 
# gjennomsnittlige feilen er i forhold til den gjennomsnittlige prisen for markedskampanjen (target) i prosent
# Gjennomsnitt og standardavvik for mean absolute error og mean squared error
print(f'Gj.snitt for mean absolute error: {np.mean(cv_mean_absolute_error):.4f}')
print(f'St.avvik for mean absolute error: {np.std(cv_mean_absolute_error):.4f}')
print(f'Gj.snitt for mean square error: {np.mean(cv_mean_square_error):.4f}')
print(f'St.avvik for mean square error: {np.std(cv_mean_square_error):.4f}')

# Beregn gjennomsnittelig feil
cv_mean_absolute_error_percent = (np.mean(cv_mean_absolute_error) / y_train.mean()) * 100
cv_mean_square_error_percent = (np.mean(cv_mean_square_error)) / y_train.mean() * 100
print(f'\nGj.snitt feil for mean absolute error: {cv_mean_absolute_error_percent:.4f}')
print(f'Gj.snitt feil for mean square error: {cv_mean_square_error_percent:.4f}')
```

**Svar:**

Ettersom standardavviket er noe lavt for begge verdiene, er modellen noe stabilt.
Gjennomsnittet for feil er nogenlunde stort, som kan føre til en større feilmargin for prediksjoner.

#### Oppgave 1.5) 
Tren modellen en siste gang på alle treningsdataene og gjør prediksjon på testdataene (bruk de optimale parametrene fra oppgave 1.3)
* Hvor stor er den gjennomsnittlige feilen på prediksjonen sammenlignet med gjennomsnittlig faktisk pris i testdataene? 

```python
#TODO: Tren modellen på alle treningsdata og utfør prediksjon på testdataene
# Opprett model igjen basert på de beste parameterene fra 1.3
opt_model = DecisionTreeRegressor(
    random_state = 42,
    max_depth = 10,
    min_samples_split = 2,
    min_samples_leaf = 10)

opt_model.fit(X_train, y_train)
prediction = opt_model.predict(X_test)

# Gjennomsnittelig feil i forhold til testdataen
mae = mean_absolute_error(y_test, prediction)
mse = mean_squared_error(y_test, prediction)
mae_error_percent = (mae / np.mean(y_test)) * 100
mse_error_percent = (mse / np.mean(y_test)) * 100

print(f'Mean absolute error: {mae_error_percent:.4f}%')
print(f'Mean squared error: {mse_error_percent:.4f}%')
```

**Svar:**

Med en mean absolute error på ~26%, vil verdiene modellen predikerer være av med ~26% i forhold til testdataene.
Mean squared error på ~887 er en veldig høy verdi som tilsier at modellen har noen store feil i sine prediksjoner.
