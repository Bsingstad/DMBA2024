Denne oppgaven tar for seg et datasett fra Food Mart, som er ennærbutikkjede i USA. Food Mart har i dette datasettet samlet endel data i forbindelse med mediekampanjer. Din oppgave er å forutsi kostnadene for mediekampanjer i matvaremarkedet på grunnlag av funksjonene som tilbys.

```python
# Importer resten av pakkene du trenger her:
import pandas as pd
from sklearn.model_selection import train_test_split, KFold
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, mean_absolute_error
from sklearn.model_selection import cross_val_score
```

```python
# I denne oppgaven skal dere bruke følgende datasett:
df = pd.read_csv("./datasett1.csv")
# Her deler jeg datasettet inn i egenskaper 5(X) og target (y)
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
# Tilslutt deler jeg inn i et trening- og et testsett (75%/25%)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=0)
# PS: Ikke endre denne kodecellen!
```

#### Oppgave 1.3) 
Velg en modellarkitektur (gjerne fra Scikit-Learn) og gjør et strukturert hyperparametersøk. 
* Gjør et søk på minst 3 paramtere og minst 2 verdier pr parameter
* Bruk 3-folds kryssvalidering for hver parameterkombinasjon
* Velg en metrikk som passer for målet (tips: "neg_mean_squared_error" i sklearn) 
* Skriv ut den parameterkobinasjonen som ga den beste skåren, samt skåren. Kommenter resultatet

```python
#TODO: Utfør hyperparametersøk for en modell
```

```python
randForestReg = RandomForestRegressor(random_state=32)

param_grid = {
    'n_estimators': [50, 100, 200],         # Antall trær
    'max_depth': [5, 10, None],            # Maksimal dybde
    'min_samples_split': [2, 5, 10]        # Minimum prøver for splitting
}

grid_search = GridSearchCV(
    estimator=randForestReg,
    param_grid=param_grid,
    scoring='neg_mean_squared_error',  
    cv=3, 
    verbose=2,
    n_jobs=-1 
)

grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
best_score = grid_search.best_score_

print("Beste parameterne: ", best_params)
print("Beste score: ", best_score)
```

**Svar:**
Her har man funnet ut at beste depth er 10, beste min_samples er 10 og n_estimators er 200 av de verdiene som er beste verdiene for å lage en modell. Den beste modellen fikk -850 score i MSE, som kan forbedres. **Rettelse:** MSE-Score er +850.

#### Oppgave 1.4) 
Bruk hyperparameterkombinasjonen fra oppgave 1.3 og kryssvalider den modellen på nytt med K >= 5 
* Logg "mean absolute error" og "mean squared error" for hver kryssvalideringsrunde
* Oppgi gjennomsnitt og standardavvik for "mean absolute error" og "mean squared error" etter kryssvalideringen og kommenter resultatet
* Hvor stor er den gjennomsnittlige feilen på prediksjonen sammenlignet med gjennomsnittlig faktisk pris i treningsdataene? 

```python
#TODO: Utfør kryssvalidering med de optimale parameterene fra forrige oppgave og logg mean_absolute_error og mean_squared_error
```

```python
#TODO: Beregn gjennomsnitt og standard avvik for mean_absolute_error og mean_squared_error og beregn hvor stor den 
# gjennomsnittlige feilen er i forhold til den gjennomsnittlige prisen for markedskampanjen (target) i prosent
```

```python
best_params = {'n_estimators': 200, 'max_depth': 10, 'min_samples_split': 10}

randForestReg = RandomForestRegressor(
    n_estimators=best_params['n_estimators'],
    max_depth=best_params['max_depth'],
    min_samples_split=best_params['min_samples_split'],
    random_state=32
)

#Krossvalidering
kf = KFold(n_splits=5, shuffle=True, random_state=32)

#Lager scorer
scr_mae = make_scorer(mean_absolute_error, greater_is_better=False)
scr_mse = make_scorer(mean_squared_error, greater_is_better=False)

#Kalkulere verdiene for krossvalideringen
mae = cross_val_score(randForestReg, X_train, y_train, scoring=scr_mae, cv=kf)
mse = cross_val_score(randForestReg, X_train, y_train, scoring=scr_mse, cv=kf)

#Gjør tallene positivb
mae = -mae
mse = -mse

print("Mean_absolute_error: ", mae)
print("Mean_squared_error: ", mse)

#Beregne gjennomsnitt og standardavik
mean_mae = mae.mean()
std_mae = mae.std()

mean_mse = mse.mean()
std_mse = mse.std()

#Finne gjennomsnittet for prisen i target
mean_target_price = y_train.mean()

#Prosent avik
pro_mae = 100 * (mean_mae / mean_target_price)

print(f"Gjennomsnitt (MAE): {mean_mae:.2f}")
print(f"Standardavik (MAE): {std_mae:.2f}")

print(f"Gjennomsnitt (MSE): {mean_mse:.2f}")
print(f"Standardavik (MSE): {std_mse:.2f}")

print(f"Prosentavik på mae: {pro_mae:.2f}")
```

**Svar:**
Modellen viser at gjennomsnittlige absolutt feil (MAE) på 25.08 fra de faktiske verdiene, med en standardavik på 0.13. Det betyr at det er konsistent mellom foldene i kryssvalideringen.

Gjennomsnittlig kvadrat feil (MSE) på 848.63. Dette blir gjort kvadratisk så blir mye større feil siden den er mer følsom. Aviken på MSE på 10.77 er høyere, men fortsatt godkjent i denne modellen.

I denne modellen er det prosentavik på 25.24% fra de faktiske prisene. 

PS: Gjorde ingen scaling denne gangen, så tar litt tid.
