I denne oppgaven skal vi forsøke å forutsi oppsigelser ved en arbeidsplass basert på en rekke variabler som f.eks alder, kjønn, utdannelse, lønn, osv.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score,confusion_matrix,recall_score,precision_score, classification_report
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from scipy.stats import ttest_ind
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
# I denne oppgaven skal dere bruke følgende datasett:
df = pd.read_csv("./datasett2.csv")
```

### Preprosessering:
For å forenkle håndteringen av kategoriske data har jeg her skrevet noe kode som gjør om strenger til intergers

Følgende kolonner er formatert som *object* og inneholder kategoriske variabler i form av strenger

```python
kolonner_obj = df.select_dtypes(include=['object']).columns
print(kolonner_obj)
```

I koden under omformer jeg strengene til integers og gir dere "nøkkelen" til å konvertere det tilbake igjen

```python
for i in kolonner_obj:
    print("-----------")
    print(i)
    print("-----------")
    for num, j in enumerate(list(df[i].unique())):
        print("{} = {}".format(j,num))
        df[i].loc[df[i]==j] = num
    df[i] = pd.to_numeric(df[i])
```

I outputen fra koden nedenfor viser datatypen til de forskjellige kolonnene, som nå etter preprosesseringen bare består av integers (int64). 

```python
print(df.dtypes)
```

```python
# Her deler jeg datasettet inn i egenskaper (X) og target (y) (fjerner også den første kolonnen siden det bare er en ID 
# og har dermed ingen prediktiv verdi)

X = df.iloc[:,1:-1]
y = df.iloc[:,-1]

# Tilslutt deler jeg inn i et trening- og et testsett (75%/25%)

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=0)

# PS: Ikke endre denne kodecellen!
```

```python
# Model 1: Logistic Regression

# Uten skalering tok det evigheter
scaler = StandardScaler()
sc_X_train = scaler.fit_transform(X_train)

# Brukes for kryssvalidering
rfk = RepeatedKFold(n_splits=5, n_repeats=10, random_state=32)

logreg = LogisticRegression(solver='lbfgs', max_iter=10000, random_state=32)
f1_logreg = cross_val_score(logreg, sc_X_train, y_train, scoring='f1', cv=rfk)
#print("LogReg F1-Score ", f1_logreg)
print("LogReg Accuracy-Score", f1_logreg.mean())

# Model 2: 
randForest = RandomForestClassifier(random_state=32)
f1_randForest = cross_val_score(randForest, sc_X_train, y_train, scoring='f1', cv=rfk)
#print("RandForest F1-Score ", f1_randForest)
print("RandForest Accuracy-Score", f1_randForest.mean())

plt.figure(figsize=(10, 6))
plt.boxplot([f1_logreg, f1_randForest], vert=True, patch_artist=True, labels=["Logistic Regression", "Random Forest Classifier"])
plt.title("F1-scores")
plt.ylabel("F1")
plt.show()
```

```python
t, p = ttest_ind(f1_logreg, f1_randForest)
print("T-Verdi:", t)
print("P-Verdi:", p)
```

**Svar**
Med verdien på T-verdien betyr det at forskjellen på gjennomsnittet er svært høyt. LogReg hadde en accuracy score på 0.3074147097975628, som er mye høyere enn random forest classifier på 0.15163353201169386. Det betyr at LogReg var mer korrekt i dette datasettet. P-verdien sier noe om hvor tilfeldig dette resultatet ville blitt. 4.126656241555562e-17 ville si at det er statisistik "umulig" at dette var tilfeldig.

```python
#TODO: repeterte kryssvalideringer for de samme to modellene som i oppgave 2.2, men denne gangen med SMOTE-oppsampling 
#from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(random_state=32)

# Model 1 : Logistic Regression med SMOTE
logreg = Pipeline([('smote', SMOTE(random_state=32)),('logreg', LogisticRegression(solver='lbfgs', max_iter=10000, random_state=32))])
f1_logreg = cross_val_score(logreg, sc_X_train, y_train, scoring='f1', cv=rfk)
#print("LogReg F1-Score ", f1_logreg)
print("LogReg Accuracy-Score", f1_logreg.mean())

# Model 2: 
randForest = Pipeline([('smote', SMOTE(random_state=32)), ('rf', RandomForestClassifier(random_state=32))])
f1_randForest = cross_val_score(randForest, sc_X_train, y_train, scoring='f1', cv=rfk)
#print("RandForest F1-Score ", f1_randForest)
print("RandForest Accuracy-Score", f1_randForest.mean())

plt.figure(figsize=(10, 6))
plt.boxplot([f1_logreg, f1_randForest], vert=True, patch_artist=True, labels=["Logistic Regression", "Random Forest Classifier"])
plt.title("F1-scores med SMOTE")
plt.ylabel("F1")
plt.show()

```

```python
t, p = ttest_ind(f1_logreg, f1_randForest)
print("T-Verdi:", t)
print("P-Verdi:", p)
```

**Svar:**
Begge modellene ble mye mer effektive. Random Forest Classifier som fikk bare en accuracy score på 0.15, gikk til 0.32 med å gjøre settet balansert. Da slår den 0.30 som Logistic Regression ubalansert. Men logreg ble bare enda mer effektiv med på score 0.40 når den ble balansert. T-verdien ble nesten halvert fra 10,21 til 5,76, som betyr de nærmer seg mer til å være like effektiv.
