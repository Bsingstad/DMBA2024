#### Oppgave 2.2)
Velg to modeller og gjør repeterte kryssvalideringer $(10)$ med $K=5$
* Velg hvorvidt du vil bruke "default" hyperparamtre for modellene eller om du vil gjøre et hyperparametersøk
* Skår ytelsen til modellene ved hjelp av F1-score og accuracy
* Visualiser F1-skårene fra de repeterte kryssvalideringene som et boksplot
* Undersøk om modell 1 er bedre eller dårligere enn modell 2 ved å gjøre en to-sidig T-test på F1-skårene fra de repeterte kryssvalideringene 
* Tolk resultatene skriftlig til slutt 


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score, RepeatedStratifiedKFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, make_scorer
from scipy import stats

# Dummy data (Bytt ut med ditt eget datasett)
X_train = np.random.rand(100, 10)  # Eksempeldata (100 prøver, 10 funksjoner)
y_train = np.random.randint(0, 2, size=100)  # Binære målverdier (0 eller 1)

# Modeller
decision_tree_model = DecisionTreeClassifier(random_state=42)
log_regr_model = LogisticRegression(random_state=42, max_iter=1000)

# RepeatedKFold for kryssvalidering
rkf = RepeatedStratifiedKFold(n_splits=5, n_repeats=10, random_state=42)

# F1-score for Decision Tree
f1_score_decision_tree = cross_val_score(decision_tree_model, X_train, y_train, cv=rkf, scoring="f1")

# F1-score for Logistic Regression
f1_score_log_regr = cross_val_score(log_regr_model, X_train, y_train, cv=rkf, scoring="f1")

# Accuracy for Decision Tree (valgfritt for sammenligning)
accuracy_decision_tree = cross_val_score(decision_tree_model, X_train, y_train, cv=rkf, scoring="accuracy")

# Accuracy for Logistic Regression (valgfritt for sammenligning)
accuracy_log_regr = cross_val_score(log_regr_model, X_train, y_train, cv=rkf, scoring="accuracy")

# Visualisering av F1-scores som boksplot
df_f1_score = pd.DataFrame({
    "Decision Tree": f1_score_decision_tree,
    "Logistic Regression": f1_score_log_regr
})

# Boksplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_f1_score, palette="Set3")
plt.title("F1-scores for Decision Tree and Logistic Regression")
plt.ylabel("F1-score")
plt.xticks([0, 1], ['Decision Tree', 'Logistic Regression'])
plt.show()

# T-test for å sammenligne F1-scorene
t_stat, p_value = stats.ttest_rel(f1_score_decision_tree, f1_score_log_regr)

# Tolk resultatene
print("T-statistic:", t_stat)
print("P-value:", p_value)

if p_value < 0.05:
    if np.mean(f1_score_decision_tree) > np.mean(f1_score_log_regr):
        print("Decision Tree er bedre enn Logistic Regression basert på F1-score.")
    else:
        print("Logistic Regression er bedre enn Decision Tree basert på F1-score.")
else:
    print("Ingen signifikant forskjell mellom modellene.")

# Skriv ut gjennomsnittlig F1-score og accuracy for modellene
print(f"\nGjennomsnittlig F1-score for Decision Tree: {np.mean(f1_score_decision_tree)}")
print(f"Gjennomsnittlig F1-score for Logistic Regression: {np.mean(f1_score_log_regr)}")

print(f"\nGjennomsnittlig accuracy for Decision Tree: {np.mean(accuracy_decision_tree)}")
print(f"Gjennomsnittlig accuracy for Logistic Regression: {np.mean(accuracy_log_regr)}")
```

**Svar:**

Logistisk regresjon har høyere F1-score enn Decision Tree, noe som betyr at den har bedre balanse mellom presisjon og recall. Decision Tree har derimot høyere accuracy, men dette er ikke nødvendigvis et bedre mål på ytelse i tilfelle ubalanserte data.

T-testen viser at det er en signifikant forskjell mellom modellene (p-verdi = 0.00025). Logistisk regresjon har høyere gjennomsnittlig F1-score (0.6376) enn Decision Tree (0.5632), som indikerer at den presterer bedre på balansering av presisjon og recall.

Til tross for høyere accuracy for Decision Tree (0.505), viser F1-score at logistisk regresjon er mer egnet for dette datasettet.

#### Oppgave 2.3) 
Tren de samme modellene men denne gangen med et balansert datasett
* Bruk imbalaced learn og SMOTE til å oppsample minioritetsklassen (bare oppsample treningsdataene i kryssvalideringen - ikke valideringsdataene)
* Skår ytelsen til modellene ved hjelp av F1-score og accuracy
* Visualiser F1-skårene fra de repeterte kryssvalideringene som et boksplot
* Undersøk om modell 1 er bedre eller dårligere enn modell 2 ved å gjøre en to-sidig T-test på F1-skårene fra de repeterte kryssvalideringene 
* Tolk resultatene og sammenlign resultatene med 2.2) 

```python
# Importer nødvendige biblioteker
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import cross_val_score, RepeatedStratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Definer modellene
log_regr_model = LogisticRegression(random_state=42, max_iter=1000)
decision_tree_model = DecisionTreeClassifier(random_state=42)

# Repeterte kryssvalideringer
k_fold = RepeatedStratifiedKFold(n_splits=5, n_repeats=10, random_state=42)

# Funksjon for å bruke SMOTE på treningsdataene i kryssvalideringen
def smote_upsample(X_train, y_train):
    smote = SMOTE(random_state=42)
    X_res, y_res = smote.fit_resample(X_train, y_train)
    return X_res, y_res

# Skåre ytelsen til modellene: F1-score og accuracy med SMOTE oppsampling på treningsdataene
f1_score_log_regr = []
f1_score_decision_tree = []
accuracy_log_regr = []
accuracy_decision_tree = []

for train_idx, val_idx in k_fold.split(X_train, y_train):
    X_train_fold, y_train_fold = X_train[train_idx], y_train[train_idx]
    X_val_fold, y_val_fold = X_train[val_idx], y_train[val_idx]
    
    # Bruk SMOTE for å oppsample treningsdataene
    X_train_res, y_train_res = smote_upsample(X_train_fold, y_train_fold)

    # Tren modellene på de oppsamplede dataene
    log_regr_model.fit(X_train_res, y_train_res)
    decision_tree_model.fit(X_train_res, y_train_res)
    
    # Evaluer modellene på valideringsdataene
    y_pred_log_regr = log_regr_model.predict(X_val_fold)
    y_pred_decision_tree = decision_tree_model.predict(X_val_fold)

    # Lagre F1-score og accuracy for hver fold
    f1_score_log_regr.append(f1_score(y_val_fold, y_pred_log_regr))
    f1_score_decision_tree.append(f1_score(y_val_fold, y_pred_decision_tree))
    accuracy_log_regr.append(accuracy_score(y_val_fold, y_pred_log_regr))
    accuracy_decision_tree.append(accuracy_score(y_val_fold, y_pred_decision_tree))

# Visualisering av F1-score med boksplot
df_f1_score = pd.DataFrame({
    "Logistisk Regresjon": f1_score_log_regr,
    "Decision Tree": f1_score_decision_tree
})
df_f1_score.plot.box()
plt.title("F1-score for modellene etter SMOTE-oppampling")
plt.ylabel("F1-score")
plt.show()

# Gjennomsnittlige F1-scores og accuracy for begge modellene
print(f"Gjennomsnittlig F1-score for logistisk regresjon: {np.mean(f1_score_log_regr)}")
print(f"Gjennomsnittlig F1-score for decision tree: {np.mean(f1_score_decision_tree)}")
print(f"Gjennomsnittlig accuracy for logistisk regresjon: {np.mean(accuracy_log_regr)}")
print(f"Gjennomsnittlig accuracy for decision tree: {np.mean(accuracy_decision_tree)}")

# T-test for å sammenligne modellene
t_stat, p_value = stats.ttest_ind(f1_score_log_regr, f1_score_decision_tree)
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Tolk resultatene
if p_value < 0.05:
    if np.mean(f1_score_log_regr) > np.mean(f1_score_decision_tree):
        print("Logistisk regresjon er bedre enn Decision Tree basert på F1-score.")
    else:
        print("Decision Tree er bedre enn Logistisk regresjon basert på F1-score.")
else:
    print("Ingen signifikant forskjell mellom modellene.")
```

**Svar:**

Etter å ha balansert datasettet ved hjelp av SMOTE, viser resultatene at Decision Tree har bedre ytelse enn Logistisk regresjon når det gjelder F1-score. Gjennomsnittlig F1-score for Decision Tree er 0.537, mens Logistisk regresjon har 0.487. Dette indikerer at Decision Tree har bedre balanse mellom presisjon og recall og dermed håndterer det balanserte datasettet mer effektivt. Når vi ser på accuracy, oppnådde Decision Tree også et noe høyere gjennomsnitt (0.482) sammenlignet med Logistisk regresjon (0.448). Den to-sidige T-testen gir en T-statistikk på -2.165 og en p-verdi på 0.033, som er under signifikansnivået på 0.05. Dette bekrefter at det er en signifikant forskjell i F1-scorene mellom modellene, med Decision Tree som den overlegne modellen. Sammenlignet med resultatene fra oppgave 2.2, ser vi at SMOTE har forbedret Decision Tree betydelig, mens Logistisk regresjon fortsatt presterer dårligere i et balansert scenario.

#### Oppgave 2.4)
Tren den beste modellen på nytt og utfør prediksjon på testsettet. Mål nå ytelsen på testsettet ved bruk av sensitivitet, spesifisitet og presisjon, itillegg til F1-skår og accuracy

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
from sklearn.metrics import classification_report
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)


# Ensure X_train and X_test are DataFrames
X_train_df = pd.DataFrame(X_train, columns=[f"feature_{i}" for i in range(X_train.shape[1])])
X_test_df = pd.DataFrame(X_test, columns=[f"feature_{i}" for i in range(X_test.shape[1])])

# Ensure columns are aligned
X_test_df = X_test_df[X_train_df.columns]

# Apply SMOTE
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train_df, y_train)

# Train Decision Tree
best_model = DecisionTreeClassifier(random_state=42)
best_model.fit(X_train_smote, y_train_smote)

# Predict and calculate metrics
y_pred = best_model.predict(X_test_df)
conf_matrix = confusion_matrix(y_test, y_pred)
tn, fp, fn, tp = conf_matrix.ravel()

# Sensitivity, Specificity, etc.
sensitivity = tp / (tp + fn)
specificity = tn / (tn + fp)
precision = tp / (tp + fp)
f1 = 2 * (precision * sensitivity) / (precision + sensitivity)
accuracy = (tp + tn) / (tp + tn + fp + fn)

print(f"Sensitivitet (Recall): {sensitivity:.4f}")
print(f"Spesifisitet: {specificity:.4f}")
print(f"Presisjon: {precision:.4f}")
print(f"F1-score: {f1:.4f}")
print(f"Accuracy: {accuracy:.4f}")

print("\nKlassifiseringsrapport:")
print(classification_report(y_test, y_pred))
```

**Svar:**

Resultatene viser at modellen har en ekstrem skjevhet mot å klassifisere alle tilfeller som tilhørende klasse 1, noe som gir perfekt recall for denne klassen, men null spesifisitet og svært lav presisjon (11,9 %). Den lave F1-scoren og nøyaktigheten på 11,9 % indikerer at modellen ikke er balansert og feiler totalt på majoritetsklassen 0. Denne ubalansen skyldes sannsynligvis skjevheter i datasettet, der klassen 0 er underrepresentert. For å forbedre modellen kan datasettet balanseres ved hjelp av teknikker som SMOTE eller vekting av klasser i modellen, og evalueringen kan justeres ved å fokusere på metrikker som balansert nøyaktighet for å reflektere ytelsen bedre.


#### Oppgave 2.5)
Del opp testsettet basert på kjønn og gjør en analyse av sensitivitet, spesifisitet og presisjon for menn og det samme for kvinner. Skriv ut og diskuter resultatet og forsøk å svar på følgende spørsmål:
* Er modellen like god for begge kjønn? 
* Hvis ikke, hva tror du er årsaken og hvilke tiltak kan vi gjøre for å forbedre modellen? 

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, recall_score, precision_score
from imblearn.over_sampling import SMOTE

# Anta at data er lastet som en DataFrame
df = pd.read_csv("./datasett2.csv")

# Identifiser kategoriske kolonner og konverter til numerisk
kolonner_obj = df.select_dtypes(include=['object']).columns
for col in kolonner_obj:
    for num, unique_val in enumerate(df[col].unique()):
        df[col] = df[col].replace(unique_val, num)
    df[col] = pd.to_numeric(df[col])

# Del datasettet i funksjoner (X) og target (y)
X = df.iloc[:, 1:-1]  # Forutsatt at første kolonne er ID, siste er target
y = df.iloc[:, -1]

# Del inn i trening- og testsett
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# Bruk SMOTE for å balansere treningsdataene
smote = SMOTE(random_state=42)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)

# Tren modellen på det balanserte datasettet
best_model = DecisionTreeClassifier(random_state=42)
best_model.fit(X_train_balanced, y_train_balanced)

# Forutsetter at 'Gender' er en kolonne i datasettet og representert som 0 (mann) og 1 (kvinne)
X_test_male = X_test[X_test["Gender"] == 0]
X_test_female = X_test[X_test["Gender"] == 1]
y_test_male = y_test[X_test["Gender"] == 0]
y_test_female = y_test[X_test["Gender"] == 1]

# Utfør prediksjoner for hvert kjønn
y_pred_male = best_model.predict(X_test_male)
y_pred_female = best_model.predict(X_test_female)

# Beregn sensitivitet, spesifisitet, og presisjon for menn
conf_matrix_male = confusion_matrix(y_test_male, y_pred_male)
tn_male, fp_male, fn_male, tp_male = conf_matrix_male.ravel()
sensitivity_male = tp_male / (tp_male + fn_male)
specificity_male = tn_male / (tn_male + fp_male)
precision_male = tp_male / (tp_male + fp_male)

# Beregn sensitivitet, spesifisitet, og presisjon for kvinner
conf_matrix_female = confusion_matrix(y_test_female, y_pred_female)
tn_female, fp_female, fn_female, tp_female = conf_matrix_female.ravel()
sensitivity_female = tp_female / (tp_female + fn_female)
specificity_female = tn_female / (tn_female + fp_female)
precision_female = tp_female / (tp_female + fp_female)

# Utskrift av resultater
print("Ytelse til modellen for menn:")
print(f"Sensitivitet (Recall): {sensitivity_male:.4f}")
print(f"Spesifisitet: {specificity_male:.4f}")
print(f"Presisjon: {precision_male:.4f}")
print(f"Confusion Matrix for menn:\n{conf_matrix_male}")

print("\nYtelse til modellen for kvinner:")
print(f"Sensitivitet (Recall): {sensitivity_female:.4f}")
print(f"Spesifisitet: {specificity_female:.4f}")
print(f"Presisjon: {precision_female:.4f}")
print(f"Confusion Matrix for kvinner:\n{conf_matrix_female}")

# Diskusjon av resultatene
if abs(sensitivity_male - sensitivity_female) > 0.1 or abs(specificity_male - specificity_female) > 0.1:
    print("\nModellen presterer ulikt for menn og kvinner.")
    print("Årsaker kan være skjev fordeling i data, bias i funksjonene, eller behov for ytterligere balansering.")
else:
    print("\nModellen presterer likt for begge kjønn.")

# Forbedringsforslag
print("\nFor å forbedre modellen kan vi:")
print("- Analysere funksjonenes betydning for prediksjon.")
print("- Bruke avanserte metoder som ensemble-modeller.")
print("- Balanser datasettet bedre ved å bruke SMOTE på hele datasettet eller vektjustering i modellen.")
```

**Svar:**

Resultatet viser at modellen har omtrent lik ytelse for begge kjønn i sensitivitet (recall) og spesifisitet, men en tydelig forskjell i presisjon, der den presterer bedre for menn enn kvinner. Mulig årsak: Datasettet kan være skjevt fordelt, med flere menn enn kvinner, noe som kan gi modellen en fordel på menns testdata. Tiltak for forbedring: Bruk av SMOTE eller lignende teknikker for å balansere datasettet før oppdeling. Utforsk modeller som håndterer ubalanserte data bedre, som Random Forest med class_weight='balanced'. Analysér funksjonenes innvirkning for å sikre at viktige trekk er jevnt fordelt mellom kjønnene.
