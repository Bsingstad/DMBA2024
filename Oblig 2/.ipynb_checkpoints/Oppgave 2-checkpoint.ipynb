# Obligatorisk innlevering 2 - oppgave 2

I denne oppgaven skal vi forsøke å forutsi oppsigelser ved en arbeidsplass basert på en rekke variabler som f.eks alder, kjønn, utdannelse, lønn, osv.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
import numpy as np
from sklearn.model_selection import RepeatedKFold, cross_val_score
from scipy.stats import ttest_ind
from sklearn.metrics import confusion_matrix, precision_score, f1_score, accuracy_score
```

```python
# I denne oppgaven skal dere bruke følgende datasett:
df = pd.read_csv("./datasett2.csv")
```

### Preprosessering:
For å forenkle håndteringen av kategoriske data har jeg her skrevet noe kode som gjør om strenger til intergers

Følgende kolonner er formatert som *object* og inneholder kategoriske variabler i form av strenger

```python
kolonner_obj = df.select_dtypes(include=['object']).columns
print(kolonner_obj)
```

I koden under omformer jeg strengene til integers og gir dere "nøkkelen" til å konvertere det tilbake igjen

```python
for i in kolonner_obj:
    print("-----------")
    print(i)
    print("-----------")
    for num, j in enumerate(list(df[i].unique())):
        print("{} = {}".format(j,num))
        df[i].loc[df[i]==j] = num
    df[i] = pd.to_numeric(df[i])
```

I outputen fra koden nedenfor viser datatypen til de forskjellige kolonnene, som nå etter preprosesseringen bare består av integers (int64). 

```python
print(df.dtypes)
```

```python
# Her deler jeg datasettet inn i egenskaper (X) og target (y) (fjerner også den første kolonnen siden det bare er en ID 
# og har dermed ingen prediktiv verdi)

X = df.iloc[:,1:-1]
y = df.iloc[:,-1]

# Tilslutt deler jeg inn i et trening- og et testsett (75%/25%)

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=0)

# PS: Ikke endre denne kodecellen!
```

#### Oppgave 2.1) 
Visualiser $y\_train$ og $y\_val$ i et histogram. Kan du, basert på figuren, se hva slags problem vi skal løse (regresjon/klassifisering)? Begrunn svaret

```python
#TODO: visualiser y_train og y_test i et histogram
plt.hist(y_train, edgecolor='black', label='y_train')
plt.hist(y_test, edgecolor='black', label='y_test')
plt.legend()
plt.xlabel('Oppsigelser')
plt.ylabel('Ansatte')
plt.title('y_train og y_test')
```

**Svar:**

Siden det er kun to mulige verdier, 0 og 1, er det et klassifiseringsproblem.

#### Oppgave 2.2)
Velg to modeller og gjør repeterte kryssvalideringer $(10)$ med $K=5$
* Velg hvorvidt du vil bruke "default" hyperparamtre for modellene eller om du vil gjøre et hyperparametersøk
* Skår ytelsen til modellene ved hjelp av F1-score og accuracy
* Visualiser F1-skårene fra de repeterte kryssvalideringene som et boksplot
* Undersøk om modell 1 er bedre eller dårligere enn modell 2 ved å gjøre en to-sidig T-test på F1-skårene fra de repeterte kryssvalideringene 
* Tolk resultatene skriftlig til slutt 


```python
#TODO: repeterte kryssvalideringer for to modeller
# Velger decision tree og random forest, begge bruker default parametere
rf = RandomForestClassifier(random_state = 42)
dt = DecisionTreeClassifier(random_state = 42)

cv = RepeatedKFold(n_splits = 5, n_repeats = 10, random_state = 42)

# Evaluer modellene
rf_f1_score = cross_val_score(rf, X_train, y_train, cv = cv, scoring='f1')
rf_accuracy_score = cross_val_score(rf, X_train, y_train, cv = cv, scoring='accuracy')
dt_f1_score = cross_val_score(dt, X_train, y_train, cv = cv, scoring = 'f1')
dt_accuracy_score = cross_val_score(dt, X_train, y_train, cv = cv, scoring = 'accuracy')

print(f'Gj.snitt F1-score for Random Forest: {np.mean(rf_f1_score):.4f}')
print(f'Gj.snitt accuracy for Random Forest: {np.mean(rf_accuracy_score):.4f}')
print(f'Gj.snitt F1-score for Decision Tree: {np.mean(dt_f1_score):.4f}')
print(f'Gj.snitt accuracy for Decision Tree: {np.mean(dt_accuracy_score):.4f}')

# Lag boxplot av F1-scorene
plt.boxplot([rf_f1_score, dt_f1_score], labels = ['Random Forest', 'Decision Tree'], whis = 1.5)
plt.title('F1 Score mellom Decision Tree og Random Forest')
plt.ylabel('F1 Score')

# Utfør en to-sidig T-test på F1-scorene.
# _ betyr en throwaway variabel.
_, p_value = ttest_ind(rf_f1_score, dt_f1_score)

alpha = 0.05
if p_value < alpha:
    if np.mean(rf_f1_score) > np.mean(dt_f1_score):
        print('Random Forest er en bedre modell.')
    else:
        print('Decision Tree er en bedre modell.')
else:
    print('Det er ingen signifikant forskjell mellom modellene.')
```

**Svar:**

Basert på utskrift av gjennomsnitt har Decision Tree en høyere verdi enn Random Forest. På boksplottet ser vi også at Decision Tree har en høyere F1-score enn Random Forest. Dette støttes også av den to-sidige T-testen som konkluderer med at Decision Tree er en bedre modell for dette datasettet.

#### Oppgave 2.3) 
Tren de samme modellene men denne gangen med et balansert datasett
* Bruk imbalaced learn og SMOTE til å oppsample minioritetsklassen (bare oppsample treningsdataene i kryssvalideringen - ikke valideringsdataene)
* Skår ytelsen til modellene ved hjelp av F1-score og accuracy
* Visualiser F1-skårene fra de repeterte kryssvalideringene som et boksplot
* Undersøk om modell 1 er bedre eller dårligere enn modell 2 ved å gjøre en to-sidig T-test på F1-skårene fra de repeterte kryssvalideringene 
* Tolk resultatene og sammenlign resultatene med 2.2) 

```python
#TODO: repeterte kryssvalideringer for de samme to modellene som i oppgave 2.2, men denne gangen med SMOTE-oppsampling 
from imblearn.over_sampling import SMOTE

s = SMOTE(random_state = 42)

X_resample, y_resample = s.fit_resample(X_train, y_train)

# Samme modeller som ovenfor
rf = RandomForestClassifier(random_state = 42)
dt = DecisionTreeClassifier(random_state = 42)

cv = RepeatedKFold(n_splits = 5, n_repeats = 10, random_state = 42)

# Evaluer modellene igjen
rf_f1_score = cross_val_score(rf, X_resample, y_resample, cv = cv, scoring='f1')
rf_accuracy_score = cross_val_score(rf, X_resample, y_resample, cv = cv, scoring='accuracy')
dt_f1_score = cross_val_score(dt, X_resample, y_resample, cv = cv, scoring = 'f1')
dt_accuracy_score = cross_val_score(dt, X_resample, y_resample, cv = cv, scoring = 'accuracy')

print(f'Gj.snitt F1-score for Random Forest: {np.mean(rf_f1_score):.4f}')
print(f'Gj.snitt accuracy for Random Forest: {np.mean(rf_accuracy_score):.4f}')
print(f'Gj.snitt F1-score for Decision Tree: {np.mean(dt_f1_score):.4f}')
print(f'Gj.snitt accuracy for Decision Tree: {np.mean(dt_accuracy_score):.4f}')

# Lag oppdaert boxplot av F1-scorene
plt.boxplot([rf_f1_score, dt_f1_score], labels = ['Random Forest', 'Decision Tree'], whis = 1.5)
plt.title('F1 Score mellom Decision Tree og Random Forest')
plt.ylabel('F1 Score')

# Utfør en to-sidig T-test på de nye F1-scorene.
_, p_value = ttest_ind(rf_f1_score, dt_f1_score)

alpha = 0.05
if p_value < alpha:
    if np.mean(rf_f1_score) > np.mean(dt_f1_score):
        print('Random Forest er en bedre modell.')
    else:
        print('Decision Tree er en bedre modell.')
else:
    print('Det er ingen signifikant forskjell mellom modellene.')
```

**Svar:**

Etter å ha brukt SMOTE som forbedrer ubalansert data i datasettet ved å lage syntetiske data for minoritetsklassen, er nå Random Forest en bedre modell enn Decision Tree. F1-scorene og accuracyen har gått signifikant opp etter å ha brukt SMOTE. Dette kan også sees på boksplottet, der y-aksen nå går mye høyere enn når SMOTE ikke var brukt.

#### Oppgave 2.4)
Tren den beste modellen på nytt og utfør prediksjon på testsettet. Mål nå ytelsen på testsettet ved bruk av sensitivitet, spesifisitet og presisjon, itillegg til F1-skår og accuracy

```python
#TODO: Tren den beste modellen fra 2.2 og 2.3 på alle treningsdataene og test på testdataene
# Siden oppgaven sier å trene den beste modellen fra 2.2 og 2.3 velger jeg modellen fra 2.3, 
# siden begge oppgavene fikk forskjellig svar på hvilken modell som var best

rf = RandomForestClassifier(random_state = 42) # Den beste modellen fra 2.3
rf = rf.fit(X_resample, y_resample)
predict = rf.predict(X_test)

# Bruk en confusion matrix for å regne ut ytelse
tn, fp, fn, tp = confusion_matrix(y_test, predict).ravel()
specificity = tn / (tn + fp)
sensitivity = tp / (tp + fn)
precision = precision_score(y_test, predict)
f1 = f1_score(y_test, predict)
acc = accuracy_score(y_test, predict)

print('Ytelse for Random Forest')
print(f'Spesifisitet: {specificity:.4f}')
print(f'Sensitivitet: {sensitivity:.4f}')
print(f'Presisjon: {precision:.4f}')
print(f'F1-score: {f1:.4f}')
print(f'Accuracy: {acc:.4f}')
```

#### Oppgave 2.5)
Del opp testsettet basert på kjønn og gjør en analyse av sensitivitet, spesifisitet og presisjon for menn og det samme for kvinner. Skriv ut og diskuter resultatet og forsøk å svar på følgende spørsmål:
* Er modellen like god for begge kjønn? 
* Hvis ikke, hva tror du er årsaken og hvilke tiltak kan vi gjøre for å forbedre modellen? 

```python
#TODO: Gjør kjønnsbasert analyse på testdataene
# Del opp datasett
X_test_male = X_test[X_test['Gender'] == 0] # Basert på at male og females endres til integer i preprosseseringen.
X_test_female = X_test[X_test['Gender'] == 1]
y_test_male = y_test[X_test['Gender'] == 0]
y_test_female = y_test[X_test['Gender'] == 1]

# Bruk Random Forest igjen og gjør prediksjoner
rf = RandomForestClassifier(random_state = 42)
rf.fit(X_resample, y_resample)
predict_male = rf.predict(X_test_male)
predict_female = rf.predict(X_test_female)

# Regn ut verdier for menn
tn_male, fp_male, fn_male, tp_male = confusion_matrix(y_test_male, predict_male).ravel()
specificity_male = tn_male / (tn_male + fp_male)
sensitivity_male = tp_male / (tp_male + fn_male)
precision_male = precision_score(y_test_male, predict_male)

# Regn ut verdier for kvinner
tn_female, fp_female, fn_female, tp_female = confusion_matrix(y_test_female, predict_female).ravel()
specificity_female = tn_female / (tn_female + fp_female)
sensitivity_female = tp_female / (tp_female + fn_female)
precision_female = precision_score(y_test_female, predict_female)

print('Ytelse for Random Forest for verdier for kjønn')
print('Menn:')
print(f'Spesifisitet: {specificity_male:.4f}')
print(f'Sensitivitet: {sensitivity_male:.4f}')
print(f'Presisjon: {precision_male:.4f}')

print('\nKvinner:')
print(f'Spesifisitet: {specificity_female:.4f}')
print(f'Sensitivitet: {sensitivity_female:.4f}')
print(f'Presisjon: {precision_female:.4f}')
```

Basert på disse verdiene er modellen god på å predikere 'true negatives', så den klassifiserer sjeldent negative verdier som positive. Modellen sliter med sensitivitet så den kan ofte ta feil av 'true positives'. Siden presisjonen ligger rundt 50% har den rundt 50% rett når den predikerer positive verdier. Modellen presterer ganske likt for menn og kvinner, men den er noe bedre på sensitivitet for menn og spesifisitet og presisjon for kvinner. 

Det er mulig at datasettet inneholder en større mengde menn over kvinner eller motsatt.
Et mulig tiltak er å vurdere å bruke SMOTE slik som i oppgave 2.3 for å se om den balanserer de bedre.
